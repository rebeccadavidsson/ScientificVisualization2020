```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import numpy as np
from ipywidgets import interact, interactive, fixed, interact_manual
sns.set()
sns.set_style("darkgrid")
```

```python
dataset = pd.read_csv("cars.csv")
```

```python
dataset.columns
```

```python
columns = ['MPG', 'cylinders', 'horsepower', 'weigth', 'year']
d = dataset[columns]

# Compute the correlation matrix
corr = d.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(5, 5))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
sns.set_style("whitegrid")

data1 = dataset[dataset.year <= 75]
data2 = dataset[dataset.year > 75]
data1 = data1.sort_values("origin")
data2 = data2.sort_values("origin")
```

```python



relplot = sns.relplot(data=data1, x="horsepower", y="MPG", hue="cylinders", size="weigth", style="origin",
            sizes=(15, 200), palette="crest", alpha=0.77,
            height=6)
relplot.set(xlim=(40, 235))


# relplot2 = sns.relplot(data=data2, x="horsepower", y="MPG", hue="cylinders", size="weigth", style="origin",
#             sizes=(15, 200), palette="crest", alpha=0.77,
#             height=6)
# relplot2.set(xlim=(40, 235))
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm


cylinders = dataset['cylinders']
colors = [0, 0, 0, '#931a25', '#e6739f', '#931a25', '#cc0e74', 0, '#790c5a']
colorlist = [colors[i] for i in cylinders]

marker = {
    'US': 'o',
    'Japan': 'x',
    'Europe': 's'
}

years = {
    70: '1970',
    71: '1971',
    72: '1972',
    73: '1973',
    74: '1974',
    75: '1975',
    76: '1976',
    77: '1977',
    78: '1978',
    79: '1979',
    80: '1980',
    81: '1981',
    82: '1982',
    83: '1983'
}


markers = [marker[i] for i in dataset['origin']]

fig = plt.figure(figsize=(9,11))

ax = Axes3D(fig)

x = dataset["horsepower"]
y = dataset["MPG"]
z = dataset["year"]
size = dataset["weigth"]
size = [(n**2)/10e5*5 for n in size]
z = [int(years[i]) for i in z]

ax.set_xlabel('Horsepower', fontsize=14)
ax.set_zlabel('MPG', fontsize=14)
ax.set_ylabel('Year', fontsize=14)
ax.set_facecolor('white')
    
for xp, yp, zp, m, color, s in zip(x, y, z, markers, colorlist, size):
    ax.scatter([xp],[zp], [yp], marker=m, color=color, alpha=0.7, s=s)
plt.show()
```
